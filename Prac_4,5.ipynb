<h2>Data Visualization</h2>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
avg_age = df['Age'].mean()
df['Age']=df['Age'].replace(np.nan,avg_age)
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
# how many survived and how many not
alive = df['Survived'].value_counts()
print("Survived Count: ",alive)
```

```python
#countplot of how many survived and how many not

plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Survived')
plt.title("How Many Survived")
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.show()
```

```python
#how many survived according to gender

alive_gender = df.groupby('Sex')['Survived'].value_counts()
print(alive_gender)
```

```python
#print the countplot who suvived grouping by age
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Sex',hue='Survived')
plt.legend(title="Survived",labels=['No','Yes'])
plt.show()
```

```python
#printning the joint plot for fare vs age

sns.jointplot(data=df,x='Fare',y='Age')
plt.show()
```

```python
sns.jointplot(data=df,x='Fare',y='Age',kind='kde')
plt.show()
```

```python
sns.jointplot(data=df,x='Fare',y='Age',kind='reg')
plt.show()
```

```python
#Printing the scatterplot for age vs fare
plt.scatter(df['Age'],df['Fare'],marker='*')
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#printning the Boxplot
sns.boxplot(data=df,x='Pclass',y='Age')
plt.show()
```

```python
sns.catplot(data=df,y='Survived',hue='Pclass',kind='bar')
```

```python
sns.swarmplot(data=df,x='Pclass',y='Age')
```

```python
sns.histplot(df['Fare'],kde=True)
```
