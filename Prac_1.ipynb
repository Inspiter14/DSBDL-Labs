<h5>2.2 Deal with missing values</h5>

1. Drop data
    * Drop the whole row
    * Drop the whole column
2.  Replace data
    * Replace it by mean
    * Replace it by frequency / mode
    * Replace it by median
    * Replace it based on other functions

```python
#dropping the column
df.drop(columns=["horsepower-binned"],axis=1,inplace=True)   

#specify col name/s, axis is set to 1 as it is column, inplace is true as we want to reflect changes into the original dataframe
#if inplace is false, it will assign changes to new dataframe
#for dropping a row, use 'df.drop(labels=[0,1],axis=0,inplace=True)' labels are row indices, axis is 0 as it is a row
```

```python
#replace with mean
avg_stroke = df['stroke'].astype("float").mean(axis=0)
#This converts the values in the 'stroke' column to floating-point numbers. 
#This step is necessary if the values are stored as strings or other non-numeric types, ensuring that they can be used in mathematical operations.

print("Average of Stroke:",avg_stroke)

#replace NaN by mean value in 'stroke' column
df['stroke'] = df['stroke'].replace(np.nan, avg_stroke)  
```

```python
#replace with mode
mode_horsepower = df['horsepower'].astype("float").mode()

mode_horsepower_value=mode_horsepower[0]

print("Mode of Horsepower: ",mode_horsepower_value)

#replace NaN by mode of 'horsepower' column
df['horsepower'] = df['horsepower'].replace(np.nan,mode_horsepower_value)
```

```python
#replace with median
median_peak_rpm = df['peak-rpm'].astype("float").median()

print("Median of Peak-RPM: ",median_peak_rpm)

#replace NaN by median of 'Peak-RPM' column
df['peak-rpm'] = df['peak-rpm'].replace(np.nan,median_peak_rpm)
```

__3. Data Normalization__

<h5>Data Normalization is the process of scaling variable, so the value ranges from 0 to 1. It is also known as 'Min-Max Scaling' </h5>

```python
min_max_scaler = MinMaxScaler()

length_2d = df['length'].values.reshape(-1,1)   #convert the panda series info 2D array for normalization

df['length_scaled'] = min_max_scaler.fit_transform(length_2d)

#to scale all the data:
#data_scaled = min_max_scaler.fit_transform(df)
```
